```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\mw865\Desktop\NTI_Project\data\processed\cleaned_data.csv')
df = df.drop('review' , axis=1)
df.head(10)
```

```python
df = df.dropna()
```

```python
from sklearn.model_selection import train_test_split

X = df['clean_review']
y = df['sentiment']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.1,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1,2))
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

model = LogisticRegression(class_weight='balanced', max_iter=10000)
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score

y_pred = model.predict(X_test_tfidf)

print(classification_report(y_test, y_pred))
```

```python
print("accuracy score: ", accuracy_score(y_test, y_pred))
```

```python
sample = ["التطبيق جميل جدا وخدمه ممتازه"]

sample_tfidf = tfidf.transform(sample)
pred = model.predict(sample_tfidf)

print(f"sentence: {sample[0]}")
print(f"prediction: {pred[0]}")
```

```python
sample = ["الخدمه وحشه جدا"]

sample_tfidf = tfidf.transform(sample)
pred = model.predict(sample_tfidf)

print(f"sentence: {sample[0]}")
print(f"prediction: {pred[0]}")
```
